```python
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

%matplotlib notebook
sns.set()
```

```python
from descwl_shear_sims.simple_sim import Sim
```

## Make A Sim from a Config

```python
with open("config.yaml", "r") as fp:
    config = yaml.safe_load(fp)

data = Sim(**config).gen_sim()
```

```python
fig, axs  = plt.subplots(nrows=3, ncols=3, figsize=(8, 8))

for band_ind, band in enumerate(data):
    for epoch_ind, obs in enumerate(data[band]):
        ax = axs[band_ind, epoch_ind]
        ax.imshow(obs.image.array)
        ax.grid(False)

plt.tight_layout()
```
